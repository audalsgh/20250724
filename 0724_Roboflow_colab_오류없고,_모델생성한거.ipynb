```python
!pip install yt-dlp  # ìœ íŠœë¸Œ ì˜ìƒì„ ë‹¤ìš´ë°›ê¸°ìœ„í•œ yt-dlp ì„¤ì¹˜

from ultralytics import YOLO
import yt_dlp
import tempfile  # ìœ íŠœë¸Œ ë§í¬ë¡œ ë‹¤ìš´ë°›ê³  ë¹ ë¥¸ ì²˜ë¦¬ë¥¼ ìœ„í•œ ì„¤ì •ì„ í•˜ëŠ” ìš©ë„

youtube_url = input("ğŸ¥ ìœ íŠœë¸Œ ì˜ìƒ ì£¼ì†Œë¥¼ ì…ë ¥í•˜ì„¸ìš”: ").strip()  # https://www.youtube.com/watch?v=54nVQ-8oUzs ì‚¬ìš©í•  ì˜ˆì •

# yt-dlp ë‹¤ìš´ë¡œë“œ ì˜µì…˜
temp_dir = tempfile.gettempdir()
ydl_opts = {
        'format': 'mp4/best[height<=480]',  # 480pë¡œ ì œí•œ (ë¹ ë¥¸ ì²˜ë¦¬)
        'outtmpl': tempfile.gettempdir() + '/temp_video.%(ext)s',
        'quiet': True,
    }

# ìœ íŠœë¸Œ ë§í¬ë¡œ ì˜ìƒ ê°€ì ¸ì˜¤ê¸°
try:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        print("â¬‡ï¸ ìœ íŠœë¸Œ ì˜ìƒ ë‹¤ìš´ë¡œë“œ ì¤‘...")
        info = ydl.extract_info(youtube_url, download=True)
        video_path = ydl.prepare_filename(info)
        print(f"âœ… ë‹¤ìš´ë¡œë“œ ì™„ë£Œ: {video_path}")
except Exception as e:
    raise RuntimeError(f"âŒ ë‹¤ìš´ë¡œë“œ ì‹¤íŒ¨: {e}")

# ğŸ” ë‚´ê°€ ìœ„ì—ì„œ ì „ì´í•™ìŠµì„ í•œ YOLOv8 ëª¨ë¸ ë¡œë“œ ë° ì¶”ë¡ 
print("ğŸ” YOLOv8 ê°ì²´ íƒì§€ ì‹¤í–‰ ì¤‘...")
model = YOLO("runs/detect/traffic-custom/weights/best.pt")

import os
from google.colab import files

# ğŸ¯ ì¶”ë¡  ìˆ˜í–‰ (ê²°ê³¼ ì €ì¥)
results = model.predict(
    source=video_path,
    save=True,
    name="video-result",  # runs/detect/video-result/ í•˜ìœ„ì— ê²°ê³¼ ì €ì¥
    conf=0.25
)

# ğŸ“¦ ê²°ê³¼ ë¹„ë””ì˜¤ ê²½ë¡œ ì¶”ì •
output_path = "runs/detect/video-result/temp_video.avi"

# ğŸ“¥ ê²°ê³¼ ì˜ìƒ ë‹¤ìš´ë¡œë“œ
if os.path.exists(output_path):
    print(f"ğŸ“ ê²°ê³¼ ì˜ìƒ ë‹¤ìš´ë¡œë“œ ì¤‘: {output_path}")
    files.download(output_path)
else:
    print("âš ï¸ ê²°ê³¼ ì˜ìƒ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê²½ë¡œë¥¼ ë‹¤ì‹œ í™•ì¸í•˜ì„¸ìš”.")
```
